```python
import easyocr
import os
import pandas as pd

reader = easyocr.Reader(['en'])
images_folder = "C:/Users/admin/Downloads/gam3a/AASTSem6/AI/Project/runs/detect/predict2/crops/license-plate/pictures"

pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)  # Show all rows

data = []
texts = ""
cleaned_text = ""
for image_file in os.listdir(images_folder):
    if not image_file.endswith(('.jpg', '.jpeg', '.png')):
        continue
    image_path = f'{images_folder}/{image_file}'
    texts = reader.readtext(image_path, detail=0)
    cleaned_text =' '.join(texts).strip('[]\ license plate | _')
    data.append({'Picture Name': image_file, 'Text': cleaned_text.lower()})

df = pd.DataFrame(data)

print(df)
```
